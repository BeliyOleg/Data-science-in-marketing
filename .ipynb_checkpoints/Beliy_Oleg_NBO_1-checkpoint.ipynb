## Модель "кредит - звонок"

```python
import pandas as pd
#считаем данные
df = pd.read_csv(r'credit_call_train.csv')
dt = pd.read_csv(r'credit_call_oot.csv')
df.head()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import MinMaxScaler

X_train = df.drop('target',axis=1)
y_train = df.target
X_test = dt.drop('target',axis=1)
y_test = dt.target

#нормализация данных
mms = MinMaxScaler()
mms.fit(X_train)
X_train = pd.DataFrame(mms.transform(X_train), columns = X_train.columns)
X_test = pd.DataFrame(mms.transform(X_test), columns = X_test.columns)

parameters = [{'solver': ['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga']},
              {'penalty':['none', 'l2']},
              {'C':[0.001, 0.01, 0.1, 1, 10, 100]}]

clf = LogisticRegression(max_iter=250)

grid_search = GridSearchCV(estimator = clf,  
                           param_grid = parameters,
                           scoring = 'roc_auc',
                           cv = 5,
                           verbose=0,
                           n_jobs=-1)

grid_search.fit(X_train,y_train)
best_logreg = grid_search.best_estimator_

dt['score'] = best_logreg.predict_proba(X_test)[:,1]
dt.head()
```

```python
dt.sort_values(by='score',inplace=True)
dt['bucket'] = pd.qcut(dt['score'],10,labels=False)+1
```

```python
dt.groupby('bucket')[['target','score']].mean().plot(kind='bar')
```

```python
ds=dt.groupby('bucket')[['target','score']].mean()
ds
```

```python
import numpy as np
ece=0
for i in range(10):
    ece+=dt.loc[dt.bucket==(i+1)].shape[0]*np.abs(ds.iloc[i,0]-ds.iloc[i,1])
ece /= dt.shape[0]
print('ece =',ece)
```

```python
df['score'] = best_logreg.predict_proba(X_train)[:,1]
df.tail(10)
```

```python
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt
fpr_train,tpr_train,tresholds_train = roc_curve(df.target,df.score)
roc_auc_train = auc(fpr_train,tpr_train)
fpr_test,tpr_test,tresholds_test = roc_curve(dt.target,dt.score)
roc_auc_test = auc(fpr_test,tpr_test)
plt.figure()
plt.plot(fpr_train,tpr_train, label='AUC_ROC Train %0.2f'%(roc_auc_train))
plt.plot(fpr_test,tpr_test, label='AUC_ROC Test %0.2f'%(roc_auc_test))
plt.plot([0,1],[0,1],'k--')
plt.xlim([0.0,1.0])
plt.ylim([0.0,1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC curve')
plt.legend()
```

```python
from sklearn.metrics import precision_recall_curve
precision_train, recall_train, thresholds_train = precision_recall_curve(df.target,df.score)
pr_rec_train = auc(recall_train,precision_train)
precision_test, recall_test, thresholds_test = precision_recall_curve(dt.target,dt.score)
pr_rec_test = auc(recall_test,precision_test)
plt.figure()
plt.plot(recall_train,precision_train, label='AUC_PR Train %0.2f'%(pr_rec_train))
plt.plot(recall_test,precision_test, label='AUC_PR Test %0.2f'%(pr_rec_test))
#plt.plot([0,1],[0,1],'k--')
plt.xlim([0.0,1.0])
plt.ylim([0.0,1.05])
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision Recall curve')
plt.legend()
```

Калибровка модели

```python
import lightgbm
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
df = pd.read_csv(r'credit_call_train.csv')
dt = pd.read_csv(r'credit_call_oot.csv')
```

```python
X_train = df.drop('target',axis=1).values
y_train = df.target.values
X_test = dt.drop('target',axis=1).values
y_test = dt.target.values
```

```python
#
# Create training and validation sets
#
X_train, X_valid, y_train, y_valid = train_test_split(X_train, y_train,
                                                 test_size=0.5, stratify = y_train)

#
# Create the LightGBM data containers
#
train_data = lightgbm.Dataset(X_train, label=y_train)
test_data = lightgbm.Dataset(X_test, label=y_test)
valid_data = lightgbm.Dataset(X_valid, label=y_valid)


#
# Train the model
#

parameters = {
    'application': 'binary',
    'objective': 'binary',
    'metric': 'auc',
    'is_unbalance': 'true',
    'boosting': 'gbdt',
    'num_leaves': 31,
    'feature_fraction': 0.5,
    'bagging_fraction': 0.5,
    'bagging_freq': 20,
    'learning_rate': 0.1,
    'verbose': -1
}

model_xgb = lightgbm.train(parameters,
                       train_data,
                       valid_sets=valid_data,
                       num_boost_round=5000,
                          )
```

```python
import xgboost as xgb
def transform(x):
    np.place(x, x==1, 0.9999)
    return np.log(x/(1-x))
```

```python
tree_model = DecisionTreeClassifier(min_samples_leaf=1700,max_depth=5)
tree_model.fit(X_train,y_train)
```

```python
TREE = tree_model.tree_
indexes = TREE.apply(X_valid.astype(np.float32))
predicts_from_xgboost = model_xgb.predict(X_valid)
predicts_from_xgboost = predicts_from_xgboost.reshape((-1,1))
```

```python
from tqdm import tqdm_notebook
log_reg_dict = {}
nodes = np.unique(indexes)
model = LogisticRegression()
for node in nodes:
    model.fit(transform(predicts_from_xgboost[indexes==node]),y_valid[indexes==node])
    log_reg_dict[node] = model
```

```python
indexes_test = TREE.apply(X_test.astype(np.float32))
predicts_from_xgboost_test = model_xgb.predict(X_test)
predicts_from_xgboost_test = predicts_from_xgboost_test.reshape((-1,1))
predicts_calibrated = np.zeros_like(predicts_from_xgboost_test)
```

```python
for node in log_reg_dict.keys():
    predicts_calibrated[indexes_test==node] = log_reg_dict[node].\
        predict_proba(transform(predicts_from_xgboost_test[indexes_test==node]))[:,1].reshape((-1,1))
```

```python
dt['score'] = predicts_calibrated
dt['idx'] = dt.index
dt.head()
```

```python
dt.sort_values(by='score',inplace=True)
dt['bucket'] = pd.qcut(dt['score'],10,labels=False)+1
```

```python
dt.groupby('bucket')[['target','score']].mean().plot(kind='bar')
```

```python
ds=dt.groupby('bucket')[['target','score']].mean()
ds
```

```python
ece=0
for i in range(10):
    ece+=dt.loc[dt.bucket==(i+1)].shape[0]*np.abs(ds.iloc[i,0]-ds.iloc[i,1])
ece /= dt.shape[0]
print('ece =',ece)
```

```python
indexes_train = TREE.apply(df.drop('target',axis=1).values.astype(np.float32))
predicts_from_xgboost_train = model_xgb.predict(df.drop('target',axis=1).values)
predicts_from_xgboost_train = predicts_from_xgboost_train.reshape((-1,1))
predicts_calibrated = np.zeros_like(predicts_from_xgboost_train)
```

```python
for node in log_reg_dict.keys():
    predicts_calibrated[indexes_train==node] = log_reg_dict[node].\
        predict_proba(transform(predicts_from_xgboost_train[indexes_train==node]))[:,1].reshape((-1,1))
```

```python
df['score'] = predicts_calibrated
df.tail(10)
```

```python
fpr_train,tpr_train,tresholds_train = roc_curve(df.target,df.score)
roc_auc_train = auc(fpr_train,tpr_train)
fpr_test,tpr_test,tresholds_test = roc_curve(dt.target,dt.score)
roc_auc_test = auc(fpr_test,tpr_test)
plt.figure()
plt.plot(fpr_train,tpr_train, label='AUC_ROC Train %0.2f'%(roc_auc_train))
plt.plot(fpr_test,tpr_test, label='AUC_ROC Test %0.2f'%(roc_auc_test))
plt.plot([0,1],[0,1],'k--')
plt.xlim([0.0,1.0])
plt.ylim([0.0,1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC curve')
plt.legend()
```

```python
precision_train, recall_train, thresholds_train = precision_recall_curve(df.target,df.score)
pr_rec_train = auc(recall_train,precision_train)
precision_test, recall_test, thresholds_test = precision_recall_curve(dt.target,dt.score)
pr_rec_test = auc(recall_test,precision_test)
plt.figure()
plt.plot(recall_train,precision_train, label='AUC_PR Train %0.2f'%(pr_rec_train))
plt.plot(recall_test,precision_test, label='AUC_PR Test %0.2f'%(pr_rec_test))
#plt.plot([0,1],[0,1],'k--')
plt.xlim([0.0,1.0])
plt.ylim([0.0,1.05])
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision Recall curve')
plt.legend()
```
